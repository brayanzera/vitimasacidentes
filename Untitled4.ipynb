```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('cat_vitimas.csv',sep=";")
df
```

```python
pedestres = df[df['sit_vitima'] == 'PEDESTRE']
```

```python
contagem_por_sexo = pedestres['sexo'].value_counts()
```

```python
print(contagem_por_sexo)
```

```python
plt.bar(contagem_por_sexo.index, contagem_por_sexo.values)
```

```python
df[['dt', 'tmp']]=df['data'].str.split(' ', expand=True)
```

```python
df[['hr', 'tmp']]=df['hora'].str.split('.', expand=True)
```

```python
df['tmp']=df['dt']+' '+df['hr']
```

```python
df['datahora']=pd.to_datetime(df['tmp'])
```

```python
df_filtrado = df[df['datahora'].dt.hour.between(00, 23)]
```

```python
df_filtrado
```

```python
count_by_hour = df_filtrado['datahora'].dt.hour.value_counts().sort_index()
```

```python
count_by_hour.plot(kind='bar', title='Ocorrências por Hora (00h-23h)')
plt.xlabel('Hora')
plt.ylabel('Número de Ocorrências')
plt.xticks(rotation=0)
```

```python
plt.figure(figsize=(6, 6))
plt.pie(count_by_hour, labels=count_by_hour.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição das Ocorrências por Hora (00h-23h)')
```
